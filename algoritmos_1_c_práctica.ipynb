### Ejemplo de operaciones con numeros enteros

```python
#Ejemplo de operaciones con enteros
un_entero=1
otro_entero=2

suma_entera=un_entero + otro_entero         #Operador Suma

division_entera=un_entero // otro_entero    #Operador division entera

division= un_entero / otro_entero           #Operador division

modulo = un_entero % otro_entero            #Operador Módulo


print(f'suma_entera={suma_entera}')
print(f'division_entera = {division_entera}')
print(f'division = {division}')
print(f'modulo = {modulo}')
```

### Ejercicio: Convertir Fahrenheit a Celsius

C = (5/9)(F-32)

Para iniciar, utilizaremos la función básica "print" y la manera de agregar comentarios al código:

```python
# Formula para convertir Farenheit a Celsius
print('C = (5/9)(F-32)')
```

La función **print** no evalua expresiones, solo permite imprimir resultados por pantalla.
El siguiente paso es incluir el uso de variables para almacenar los valores numéricos:

```python
# Formula para convertir Farenheit a Celsius

fahr = 80
cel = 5//9 * fahr-32
print("fahr=", fahr, " -> cel=", cel)
```

En este caso utilizamos la función **print** para presentar el resultado de una operación, pero NO es el resultado que esperamos, esto se debe a que utilizamos el operador "//" que devuelve la parte entera de la división con redondeo al piso (floor).
En un paso siguiente vamos a utilizar el operador correcto:

```python
# Formula para convertir Farenheit a Celsius

fahr = 80
cel = 5/9 * fahr-32
print("fahr=", fahr, " -> cel=", cel)
```

Ahora tenemos un resultado diferente pero sigue sin ser correcto, en este caso se debe a la precedencia de los operadores matemáticos. Para controlar el orden en que se deben ejecutar las operaciones es necesario que incluyamos el uso de los paréntesis.

```python
# Formula para convertir Farenheit a Celsius

fahr = 80
cel = (5/9) * (fahr-32)
print("azul", fahr, " -> cel=", cel)
```

Finalmente tenemos el valor correcto pero podemos hacer algo más para presentar el reusltado mejor, en este caso vamos a utilizar otra función básica que provee Python para redondear los valores decimales:

```python
# Formula para convertir Farenheit a Celsius

fahr = 80
cel = (5/9) * (fahr-32)
print("fahr=", fahr, " -> cel=", round(cel, 2))
```

### Estructura "if" (Sentencias Condicionales)
 Es la estructura básica que le permite al programador ver si se cumple una condición, y realizar distintas acciones según si eso sucede o no.

**if** (condicion):
    
    ... lo que se hace si se cumple la condición
    
**elif** (condicion2):

    ... lo que se hace si se cumple la condicion2
     
**else**:

    ... lo que se hace si no se cumple la condición
    
    
 Siempre se comienza con **if**, después puede haber cualquier cantidad de **elif** (o ninguno) y luego se puede terminar con **else** (o no). Luego del **else** no puede haber más **elif**.


```python
a = 5
b = 9
c = a * b

if a > b:
    print('a mayor que b')
elif a < c:
    print('c es mayor que todos')
else:
    print('final')
```

```python
type(int(opcion))
```

**while** (Condicion):

**else** :

Generar una tabla de conversión de grados
Fahrenheit a Celsius partiendo de 0 hasta
100 a intervalos de 10.

```python
# Conversor fahrenheit a celsius

fahr = 0
while fahr < 101:
	cel = (5*(fahr-32))/9
	print("fahr=", round(fahr,2), "-> cel=", round(cel,2))
	fahr = fahr + 10
```

Generar una tabla de conversión de grados
Fahrenheit a Celsius partiendo de 0 hasta
100 a intervalos de 10.
Y luego de 100 a 110 a intervalos de 1.

```python
# conversor fahrenheit a celsius

fahr = 0
while fahr < 101:
	cel = (5*(fahr-32))/9
	print("fahr=", round(fahr,2), "-> cel=", round(cel,2))
	fahr = fahr + 10

fahr = 101
while fahr < 111:
	cel = (5*(fahr-32))/9
	print("fahr=", round(fahr,2), "-> cel=", round(cel,2))
	fahr = fahr + 1
```

### Estructura "for" (Ciclos o bucles)
La estructura del for permite realizar una serie de acciones con una variable, que cambia su valor en cada repetición (iteración).

for (variable) in (conjunto de valores):

... Lo que quieras hacer con tu variable
Veámoslo en un ejemplo de codigo:

```python
for x in range(0,18,6):
    print(x)
```

range(a,b,d) nos genera los valores desde a hasta b con incrementos de d, por ejemplo el siguiente codigo generara los numeros pares entre 0 y 10:

```python
for x in range(0,10,2):
    print(x)
```

Noten que el límite superior del rango nunca es incluído.

Al utilizar la instrucción range( ) no es necesario proporcionar siempre los 3 parametros. Observen que sucede cuando se proporcionan 1, 2 o 3 parámetros:

```python
print("range(5):")
for x in range(5):
    print(x)

print("range(10, 15):")
for x in range(10, 15):
    print(x)

print("range(10, 20, 3):")
for x in range(10, 20, 3):
    print(x)
```
